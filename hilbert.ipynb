```python
def fft(x):
    N = len(x)
    if N <= 1:
        return x
    even = fft(x[0::2])
    odd = fft(x[1::2])
    T = [np.exp(-2j * np.pi * k / N) * odd[k] for k in range(N // 2)]
    return [even[k] + T[k] for k in range(N // 2)] + [even[k] - T[k] for k in range(N // 2)]

def ifft(X):
    N = len(X)
    if N <= 1:
        return X
    even = ifft(X[0::2])
    odd = ifft(X[1::2])
    T = [np.exp(2j * np.pi * k / N) * odd[k] for k in range(N // 2)]
    return [(even[k] + T[k]) / 2 for k in range(N // 2)] + [(even[k] - T[k]) / 2 for k in range(N // 2)]

def hilbert_transform(signal):
    # Compute the FFT of the input signal
    N = len(signal)
    X = fft(signal)

    # Create the filter: 1 for positive frequencies, 0 for negative frequencies
    H = np.zeros(N)
    if N % 2 == 0:
        H[0] = 1
        H[1:N//2] = 2
        H[N//2] = 1
    else:
        H[0] = 1
        H[1:(N+1)//2] = 2
    
    # Apply the filter to the FFT of the signal
    X_filtered = X * H

    # Compute the inverse FFT to get the analytic signal
    analytic_signal = ifft(X_filtered)

    return analytic_signal


def unwrap_phase(phase):
    unwrapped_phase = np.copy(phase)
    N = len(phase)
    for i in range(1, N):
        delta = phase[i] - phase[i - 1]
        if delta > np.pi:
            unwrapped_phase[i:] -= 2 * np.pi
        elif delta < -np.pi:
            unwrapped_phase[i:] += 2 * np.pi
    return unwrapped_phase
```

```python
#include <stdio.h>
#include <stdlib.h>
#include <math.h>
#include <complex.h>

// Define complex type
typedef double complex cplx;

// Helper function to print complex array (for debugging)
void print_array(cplx *array, int N) {
    for (int i = 0; i < N; i++) {
        printf("(%f, %f)\n", creal(array[i]), cimag(array[i]));
    }
}

// Recursive FFT function
void fft(cplx *buf, cplx *out, int N, int step) {
    if (step < N) {
        fft(out, buf, N, step * 2);
        fft(out + step, buf + step, N, step * 2);

        for (int i = 0; i < N; i += 2 * step) {
            cplx t = cexp(-I * M_PI * i / N) * out[i + step];
            buf[i / 2] = out[i] + t;
            buf[(i + N) / 2] = out[i] - t;
        }
    }
}

// Recursive IFFT function
void ifft(cplx *buf, cplx *out, int N, int step) {
    if (step < N) {
        ifft(out, buf, N, step * 2);
        ifft(out + step, buf + step, N, step * 2);

        for (int i = 0; i < N; i += 2 * step) {
            cplx t = cexp(I * M_PI * i / N) * out[i + step];
            buf[i / 2] = (out[i] + t) / 2;
            buf[(i + N) / 2] = (out[i] - t) / 2;
        }
    }
}

// Hilbert transform
void hilbert_transform(cplx *signal, cplx *result, int N) {
    cplx *X = malloc(N * sizeof(cplx));
    cplx *H = malloc(N * sizeof(cplx));
    cplx *X_filtered = malloc(N * sizeof(cplx));
    cplx *analytic_signal = malloc(N * sizeof(cplx));

    for (int i = 0; i < N; i++) {
        X[i] = signal[i];
    }

    fft(X, result, N, 1);

    for (int i = 0; i < N; i++) {
        if (i == 0 || i == N / 2) {
            H[i] = 1;
        } else if (i < N / 2) {
            H[i] = 2;
        } else {
            H[i] = 0;
        }
    }

    for (int i = 0; i < N; i++) {
        X_filtered[i] = result[i] * H[i];
    }

    ifft(X_filtered, analytic_signal, N, 1);

    for (int i = 0; i < N; i++) {
        result[i] = analytic_signal[i];
    }

    free(X);
    free(H);
    free(X_filtered);
    free(analytic_signal);
}

// Phase unwrapping
void unwrap_phase(double *phase, double *unwrapped_phase, int N) {
    for (int i = 0; i < N; i++) {
        unwrapped_phase[i] = phase[i];
    }

    for (int i = 1; i < N; i++) {
        double delta = phase[i] - phase[i - 1];
        if (delta > M_PI) {
            for (int j = i; j < N; j++) {
                unwrapped_phase[j] -= 2 * M_PI;
            }
        } else if (delta < -M_PI) {
            for (int j = i; j < N; j++) {
                unwrapped_phase[j] += 2 * M_PI;
            }
        }
    }
}

// Main function for testing
int main() {
    int N = 8;
    cplx signal[] = {1, 1, 1, 1, 0, 0, 0, 0}; // Example signal
    cplx result[N];

    printf("Original signal:\n");
    print_array(signal, N);

    hilbert_transform(signal, result, N);

    printf("Hilbert transform result:\n");
    print_array(result, N);

    // Example phase array for unwrapping
    double phase[] = {0, M_PI / 2, M_PI, -M_PI / 2, 0, M_PI / 2, M_PI, -M_PI / 2};
    double unwrapped_phase[N];
    unwrap_phase(phase, unwrapped_phase, N);

    printf("Unwrapped phase:\n");
    for (int i = 0; i < N; i++) {
        printf("%f\n", unwrapped_phase[i]);
    }

    return 0;
}
```

```python
#include <complex.h>
#include <math.h>
#include <stddef.h>

complex double fft(complex double *x, size_t N) {
    if (N <= 1) {
        return x[0];
    }

    complex double even[N / 2];
    complex double odd[N / 2];
    for (size_t i = 0; i < N / 2; i++) {
        even[i] = x[i * 2];
        odd[i] = x[i * 2 + 1];
    }

    complex double even_fft = fft(even, N / 2);
    complex double odd_fft = fft(odd, N / 2);

    complex double T[N / 2];
    for (size_t k = 0; k < N / 2; k++) {
        T[k] = cexp(-2.0 * M_PI * I * k / N) * odd_fft[k];
    }

    complex double result[N];
    for (size_t k = 0; k < N / 2; k++) {
        result[k] = even_fft[k] + T[k];
        result[k + N / 2] = even_fft[k] - T[k];
    }

    return result;
}

complex double ifft(complex double *X, size_t N) {
    if (N <= 1) {
        return X[0];
    }

    complex double even[N / 2];
    complex double odd[N / 2];
    for (size_t i = 0; i < N / 2; i++) {
        even[i] = X[i * 2];
        odd[i] = X[i * 2 + 1];
    }

    complex double even_ifft = ifft(even, N / 2);
    complex double odd_ifft = ifft(odd, N / 2);

    complex double T[N / 2];
    for (size_t k = 0; k < N / 2; k++) {
        T[k] = cexp(2.0 * M_PI * I * k / N) * odd_ifft[k];
    }

    complex double result[N];
    for (size_t k = 0; k < N / 2; k++) {
        result[k] = (even_ifft[k] + T[k]) / 2.0;
        result[k + N / 2] = (even_ifft[k] - T[k]) / 2.0;
    }

    return result;
}

complex double *hilbert_transform(complex double *signal, size_t N) {
    complex double X[N];
    for (size_t i = 0; i < N; i++) {
        X[i] = signal[i];
    }

    complex double X_fft = fft(X, N);

    complex double H[N];
    if (N % 2 == 0) {
        H[0] = 1.0;
        for (size_t i = 1; i < N / 2; i++) {
            H[i] = 2.0;
        }
        H[N / 2] = 1.0;
    } else {
        H[0] = 1.0;
        for (size_t i = 1; i < (N + 1) / 2; i++) {
            H[i] = 2.0;
        }
    }

    complex double X_filtered[N];
    for (size_t i = 0; i < N; i++) {
        X_filtered[i] = X_fft[i] * H[i];
    }

    complex double *analytic_signal = (complex double *)malloc(N * sizeof(complex double));
    for (size_t i = 0; i < N; i++) {
        analytic_signal[i] = ifft(X_filtered, N)[i];
    }

    return analytic_signal;
}

double *unwrap_phase(double *phase, size_t N) {
    double *unwrapped_phase = (double *)malloc(N * sizeof(double));
    for (size_t i = 0; i < N; i++) {
        unwrapped_phase[i] = phase[i];
    }

    for (size_t i = 1; i < N; i++) {
        double delta = phase[i] - phase[i - 1];
        if (delta > M_PI) {
            for (size_t j = i; j < N; j++) {
                unwrapped_phase[j] -= 2 * M_PI;
            }
        } else if (delta < -M_PI) {
            for (size_t j = i; j < N; j++) {
                unwrapped_phase[j] += 2 * M_PI;
            }
        }
    }

    return unwrapped_phase;
}

```
